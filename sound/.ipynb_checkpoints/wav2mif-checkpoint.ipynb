```python
from scipy.io.wavfile import read
import scipy.signal as sps
```

```python
Fs, data = read("5_Centimeters_Per_Second_Clip_16bit_PCM_13_Seconds.wav")
from IPython.display import Audio
Audio(data[:,0], rate=Fs)
```

```python
print(Fs, data.shape)
print(data)
```

```python
filename = "gameover"
Fs, data = read("sound_"+filename+".wav")
from IPython.display import Audio
Audio(data[:,0], rate=Fs)
```

```python
print(Fs, data.shape)
print(data)
```

```python
# Your new sampling rate
new_Fs = 8000

# Resample data
number_of_samples = round(len(data) * float(new_Fs) / Fs)
data = sps.resample(data, number_of_samples)

print(data.shape)
print(data)
Audio(data[:,0], rate=new_Fs)
```

```python
depth = data.shape[0] * data.shape[1] + 1
print(depth)

outFile = open("sound_"+filename+"_hex_rom.mif", 'w')
outFile.write("WIDTH = 16;\n")
outFile.write("DEPTH = %d;\n" %depth)
outFile.write("ADDRESS_RADIX = HEX;\n")
outFile.write("DATA_RADIX = HEX;\n")
outFile.write("CONTENT BEGIN\n")
outFile.write("\n")


for i in range(data.shape[0]):
    outFile.write("%X: %04X;\n" %(2*i+1, ( int(round(data[i,0])) & 0xffff )))
    outFile.write("%X: %04X;\n" %(2*i+2, ( int(round(data[i,1])) & 0xffff )))

outFile.write("END;\n")
outFile.close()
```

```python
depth = data.shape[0] + 1
print(depth)

outFile = open("mono_sound_"+filename+"_hex_rom.mif", 'w')
outFile.write("WIDTH = 16;\n")
outFile.write("DEPTH = %d;\n" %depth)
outFile.write("ADDRESS_RADIX = HEX;\n")
outFile.write("DATA_RADIX = HEX;\n")
outFile.write("CONTENT BEGIN\n")
outFile.write("\n")


for i in range(data.shape[0]):
    outFile.write("%X: %04X;\n" %(i+1, ( int(round(data[i,0])) & 0xffff )))

outFile.write("END;\n")
outFile.close()
```
